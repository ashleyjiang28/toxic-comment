```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import bokeh.plotting as bp
```

```python
def rem_punct(word):
    w = ""
    letter = 0
    number = 0
    for ch in word.upper():
        if ch < 'A' or ch > 'Z':
            if ch < '0' or ch > '9':
                continue
            else:
                number += 1
        else:
            letter += 1
        w += ch
    if letter < number:
        w = ""
    return w

test_list = ["abc???", "Ab?C.","Th@t", "123", "...n123", "as123df", "1800AssHole"]
for test in test_list:
    print(rem_punct(test))
```

```python
df = pd.read_csv("train.csv")
df
```

```python
df = df[df["toxic"] == 1]

di = dict()
for comment in df["comment_text"]:
    for word in comment.split():
        word = rem_punct(str(word).upper())
        if "HTTP" in word or word == "":
            continue
        if word in di:
            di[word] += 1
        else:
            di[word] = 1

di
```

```python
df = df[df["severe_toxic"] == 1]

di = dict()
for comment in df["comment_text"]:
    for word in comment.split():
        word = rem_punct(str(word).upper())
        if "HTTP" in word or word == "":
            continue
        if word in di:
            di[word] += 1
        else:
            di[word] = 1
di
```

```python
valdi = dict()
for w in sorted(di, key=di.get, reverse=True):
    if di[w] in valdi:
        valdi[di[w]] = valdi[di[w]].append(str(w))
    else:
        valdi[di[w]] = [str(w)]
valdi
```
