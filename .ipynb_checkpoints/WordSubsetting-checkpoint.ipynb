```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import bokeh.plotting as bp
```

```python
def rem_punct(word):
    w = ""
    letter = 0
    number = 0
    last = ''
    lastc = 0
    for ch in word.upper():
        if lastc > 0:
            if last == ch:
                lastc += 1
                if lastc > 2:
                    continue
            else:
                last = ch
                lastc = 1
        if ch < 'A' or ch > 'Z':
            if ch < '0' or ch > '9':
                continue
            else:
                number += 1
        else:
            letter += 1
        if lastc == 0:
            last = ch
            lastc = 1
        w += ch
    if letter <= number:
        w = ""
    return w

test_list = ["abc???", "Ab?C.","Th@t", "123", "...n123", "as123df", "1800AssHole", "CARRRRRRROT", "IIIT"]
for test in test_list:
    print(rem_punct(test))
```

```python
def search_subset(df):
    di = dict()
    for comment in df["comment_text"]:
        for word in comment.split():
            word = rem_punct(str(word).upper())
            if "HTTP" in word or word == "":
                continue
            if word in di:
                di[word] += 1
            else:
                di[word] = 1
                
    return di
```
